# Tarot Card Meanings — Starter Analysis

A quick tour of the **Tarotoo Tarot Card Meanings** dataset: all 78 tarot cards (Rider–Waite–Smith tradition) with meanings, keywords, love/career/mood/spiritual contexts, planet & zodiac associations, and yes/no values.

- Source: [github.com/Tarotoo-com/tarotoo-tarot-dataset](https://github.com/Tarotoo-com/tarotoo-tarot-dataset) · DOI: [10.5281/zenodo.21268290](https://doi.org/10.5281/zenodo.21268290) · License: MIT

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/tarotoo-tarot-card-meanings/cards.csv')
print(df.shape)
df.head()
```

## Deck structure

```python
df.groupby(['arcana']).size().rename('cards').to_frame()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 3.5))
df['suit'].fillna('major arcana').value_counts().plot.bar(ax=axes[0], color='#7b47c9', title='Cards per suit')
df['yes_no'].value_counts().plot.bar(ax=axes[1], color='#d2a17d', title='Yes / No / Maybe distribution')
plt.tight_layout()
```

## Look up a card

```python
def card(name):
    row = df[df['name'].str.lower() == name.lower()].iloc[0]
    return row[['name','element','planet','zodiac','yes_no','meaning_upright','meaning_reversed','love','career','mood','spiritual']]

card('The Fool')
```

## Planet & zodiac associations

```python
df[df['arcana']=='major'][['name','planet','zodiac']].reset_index(drop=True)
```

## Most common keywords

```python
from collections import Counter
kw = Counter()
for col in ['keywords_upright','keywords_reversed']:
    for cell in df[col]:
        kw.update(k.strip() for k in str(cell).split(';'))
pd.Series(dict(kw.most_common(15))).sort_values().plot.barh(figsize=(8,5), color='#42266c', title='Top keywords across the deck')
plt.tight_layout()
```
